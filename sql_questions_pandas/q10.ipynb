### Get company name which has consistent increasing revenue

```python
import pandas as pd
import numpy as np
```

```python
revenue = pd.read_csv('./data/company_revenue.csv')
```

```python
revenue.sort_values(['company','year'], inplace=True)
revenue
```

```python
revenue['following_revenue'] = revenue.groupby('company')['revenue'].shift(1)
revenue['increasing'] = ((revenue['revenue']-revenue['following_revenue'])>0).replace({True:1,False:0})
revenue
```

```python
revenue.groupby('company').filter(lambda grp: grp['following_revenue'].count()==grp['increasing'].sum())['company'].unique()
```
