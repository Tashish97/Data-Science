### Find top 4 numbers from any 2 sections, fulfilling criterias
Problem statement : we have a table which stores data of multiple sections. every section has 3 numbers  
we have to find top 4 numbers from any 2 sections(2 numbers each) whose addition should be maximum  
so in this case we will choose section b where we have 19(10+9) then we need to choose either C or D  
because both has sum of 18 but in D we have 10 which is big from 9 so we will give priority to D.

```python
import pandas as pd
```

```python
section_data = pd.read_csv('./data/section_data.csv')
section_data.sort_values(['section','number'],ascending=[True,False],inplace=True)
section_data
```

```python
section_data['lag_numer'] = section_data.groupby('section')['number'].shift(-1)
section_data['lag_sum'] =  section_data['number'] + section_data['lag_numer']
section_data
```

```python
section_data = section_data[~section_data['lag_sum'].isna()]
section_data
```

```python
section_data.loc[:,['large_num']] = section_data.groupby('section')['number'].transform('max')
section_data
```

```python
section_data.loc[:,['section_sum']] = section_data.groupby('section')['lag_sum'].transform('max')
section_data
```

```python
section_data.loc[:,['section_rnk']] = section_data['section_sum'].rank(method='dense',ascending=False)
section_data
```

```python
section_data.loc[:,['num_rnk']] = section_data.groupby('section_sum')['large_num'].rank(method='dense',ascending=False)
section_data
```

```python
section_data.loc[(section_data['section_rnk']<=2)&(section_data['num_rnk']==1),['section','number']]
```
