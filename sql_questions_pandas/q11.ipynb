### People relationships

```python
import pandas as pd
import numpy as np
```

```python
relations = pd.read_csv('./data/relations.csv')
relations
```

```python
people  =pd.read_csv('./data/people.csv')
people
```

```python
relations.merge(people, right_on='id',left_on='c_id').merge(people,left_on='p_id',right_on='id').apply(lambda df: 0 if df['gender_y']=='M' else df['name_y'], axis=1)
```

```python
merged = relations.merge(people, right_on='id',left_on='c_id').merge(people,left_on='p_id',right_on='id').\
    assign(fathers_name = lambda df: df.apply(lambda df: df['name_y'] if df['gender_y'].strip()=='M' else None, axis=1)).\
        assign(mothers_name = lambda df: df.apply(lambda df: df['name_y'] if df['gender_y'].strip()=='F' else None, axis=1))\
            .groupby('name_x').agg({'fathers_name':pd.Series.unique,'mothers_name':pd.Series.unique})
merged
```

```python
merged['fathers_name'] = merged['fathers_name'].apply(lambda x: x[0])
merged['mothers_name'] = merged['mothers_name'].apply(lambda x: x[1])
merged.reset_index(inplace=True)
merged
```
