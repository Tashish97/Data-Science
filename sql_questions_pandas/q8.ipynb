### Pairing child and adult.
### Each child must be paired with an adult so that they can ride a fair.
### Adult can ride the fair alone but a child can't

```python
import pandas as pd
import numpy as np
```

```python
family = pd.read_csv('./data/family.csv')
family['number'] = family['person'].str.replace(pat='\D',repl='',regex=True)
family
```

```python
adults = family.loc[family['type']=='Adult']
adults
```

```python
childs = family.loc[family['type']=='Child']
childs
```

```python
pd.merge(adults,childs,on='number',how='left')[['person_x','person_y']].rename(columns={'person_x':'Adult','person_y':'Child'})
```
