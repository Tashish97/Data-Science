### Qualify candidates based on conditions.  
Find the people qualifies where each candidate qualifies both the criteria's i.e(Y and Y)  
and within their group there exixts at least one other team member i.e( from each team at least two members qualify)

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./data/ameriprise_llc.csv')
data
```

```python
data['candidate_qualify'] = (data['Criteria1']=='Y')&(data['Criteria2']=='Y')
data['candidate_qualify'] = data['candidate_qualify'].apply(lambda x: 1 if x ==True else 0)
```

```python
data_qualified = data.loc[data['candidate_qualify']==1]
data_qualified = data_qualified.drop(columns=['candidate_qualify'])
data_qualified
```

```python
data_qualified.groupby('teamID').filter(lambda grp: grp['teamID'].count()>1)
```
