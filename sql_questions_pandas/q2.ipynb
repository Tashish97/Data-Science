### Find number of gold medals for winners, who only won gold medals. 

```python
import pandas as pd
import numpy as np
```

```python
events = pd.read_csv('./data/olympic_events.csv')
```

```python
events
```

```python
merged_data = pd.merge(events,events,left_on='GOLD',right_on='SILVER',how='left').merge(events,left_on='GOLD_x',right_on='BRONZE',how='left')
merged_data
```

```python
merged_data[(merged_data['ID_y'].isna())&(merged_data['ID'].isna())].groupby('GOLD_x')['GOLD_x'].count()
```
